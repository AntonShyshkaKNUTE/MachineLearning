1. Створіть одновимірний масив (вектор) з добутками квадратів кожного числа та його подвійного значення, починаючи з 1, та виведіть його значення.

2. Створіть двовимірний масив (матрицю) розміром 3x3, заповніть його значеннями добутків косинусів кожного числа та його попереднього, починаючи з 1, та виведіть його значення.

```python
matrix = np.zeros((3, 3))

for i in range(3):
    for j in range(3):
        matrix[i, j] = np.cos((i + 1) * (j + 1))

print(matrix)
```

3. Створіть масив розміром 5x5, заповніть його випадковими цілими числами від -10 до 10 та знайдіть відсоток чисел, які мають останню цифру 5.

```python
random_array = np.random.randint(-10, 11, size=(5, 5))

numbers_with_last_digit_5 = random_array[random_array % 10 == 5]
percentage = (len(numbers_with_last_digit_5) / random_array.size) * 100

print(f"Масив:\n{random_array}\n")
print(f"Відсоток чисел з останньою цифрою 5: {percentage}%")
```

5. Створіть два одновимірних масиви розміром 8, заповніть їх випадковими цілими числами в діапазоні від 3 до 8 та виконайте на них поелементні операції додавання, віднімання та множення.

```python
array1 = np.random.randint(3, 9, size=8)
array2 = np.random.randint(3, 9, size=8)

addition_result = array1 + array2
subtraction_result = array1 - array2
multiplication_result = array1 * array2

print("Масив 1:", array1)
print("Масив 2:", array2)
print("Результат додавання:", addition_result)
print("Результат віднімання:", subtraction_result)
print("Результат множення:", multiplication_result)
```

6. Створіть два вектори розміром 4, заповніть їх довільними дійсними числами та знайдіть їх скалярний добуток.

7. Створіть дві матриці розміром 3x2 та 2x4, заповніть їх випадковими цілими числами в діапазоні від 1 до 20 та перемножте їх між собою.

```python
matrix1 = np.random.randint(1, 21, size=(3, 2))
matrix2 = np.random.randint(1, 21, size=(2, 4))

result = np.dot(matrix1, matrix2)

print("Матриця 1:")
print(matrix1)
print("\nМатриця 2:")
print(matrix2)
print("\nРезультат перемноження:")
print(result)
```

8. Створіть матрицю розміром 3x3, заповніть її випадковими цілими числами в діапазоні від 1 до 7 та знайдіть її обернену матрицю.

```python
matrix = np.random.randint(1, 8, size=(3, 3))

print("Початкова матриця:")
print(matrix)

determinant = np.linalg.det(matrix)
if determinant == 0:
    print("Не можливо знайти обернену матрицю.")
else:
    inverse_matrix = np.linalg.inv(matrix)

    print("Обернена матриця:")
    print(inverse_matrix)
```

9. Створіть матрицю розміром 5x5, заповніть її випадковими цілими числами в діапазоні від 0 до 5 та транспонуйте її.

```python
matrix = np.random.randint(0, 6, size=(5, 5))

print("Початкова матриця:")
print(matrix)

transposed_matrix = np.transpose(matrix)

print("\nТранспонована матриця:")
print(transposed_matrix)
```

10. Створіть матрицю розміром 3x4 та вектор розміром 4, заповніть матрицю випадковими дійсними числами від 0 до 1, а вектор - випадковими дійсними числами від 0 до 1, і знайдіть добуток матриці на вектор.

11. Створіть матрицю розміром 2x3 та вектор розміром 3, заповніть матрицю випадковими дійсними числами від 0 до 1, а вектор - випадковими цілими числами в діапазоні від 1 до 10, і знайдіть суму добутків елементів кожного стовпця матриці на відповідний елемент вектору.

```python
matrix = np.random.rand(2, 3)

vector = np.random.randint(1, 11, size=3)

result = np.sum(matrix * vector, axis=0)

print("Матриця:")
print(matrix)

print("\nВектор:")
print(vector)

print("\nСума добутків елементів кожного стовпця матриці:")
print(result)
```

12. Створіть дві матриці розміром 7x7, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та знайдіть відношення всіх елементів другої матриці до відповідних елементів першої матриці (поелементне ділення).

```python
matrix1 = np.random.randint(1, 11, size=(7, 7))
matrix2 = np.random.randint(1, 11, size=(7, 7))

result = matrix2 / matrix1

print("Матриця 1:")
print(matrix1)

print("\nМатриця 2:")
print(matrix2)

print("\nВідношення елементів другої матриці до елементів першої матриці:")
print(result)
```

13. Створіть дві матриці розміром 5x5, заповніть їх випадковими цілими числами в діапазоні від -10 до 10 та знайдіть суму їх елементів.

```python
matrix1 = np.random.randint(-10, 11, size=(5, 5))
matrix2 = np.random.randint(-10, 11, size=(5, 5))

sum_matrix1 = np.sum(matrix1)
sum_matrix2 = np.sum(matrix2)

print("Матриця 1:")
print(matrix1)

print("\nМатриця 2:")
print(matrix2)

print("\nСума елементів матриці 1:", sum_matrix1)
print("Сума елементів матриці 2:", sum_matrix2)
```

14. Створіть матрицю розміром 10x10, заповніть її випадковими цілими числами в діапазоні від 1 до 100 та знайдіть кількість елементів, що дорівнюють нулю.

```python
matrix = np.random.randint(1, 101, size=(10, 10))

zero_count = np.count_nonzero(matrix == 0)

print("Матриця:")
print(matrix)

print("\nКількість елементів, що дорівнюють нулю:", zero_count)
```

15. Створіть матрицю розміром 7x7, заповніть її випадковими дійсними числами в діапазоні від -1 до 1 та знайдіть вектор-стовпчик, що містить суму кубів елементів кожного стовпця матриці.

```python
matrix = np.random.uniform(-1, 1, size=(7, 7))

sum_cubed_columns = np.sum(matrix**3, axis=0)

sum_cubed_columns = sum_cubed_columns.reshape(-1, 1)

print("Матриця:")
print(matrix)

print("\nСума кубів елементів кожного стовпця:")
print(sum_cubed_columns)
```
